````python
### Start code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('data.csv')
df['Day']=pd.DatetimeIndex(df['Day']).year
df.set_index(['Day','Average temperature (°F)'],inplace=True,drop=False)
fig_dims = (15,8)
#plt.title('Average temperature across years')
fig, ax = plt.subplots(figsize=fig_dims)
plot1=sns.boxplot(x = "Day", y ="Average temperature (°F)", ax=ax, data=df).set_title('Average temperature across years')

````

````python
###Start code here

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_csv('data.csv')
#print(df1.corr())
plt.figure(figsize=(10,10))
plot2=sns.heatmap(df1.corr(),annot=True,annot_kws={'size': 6}).set_title('Correlation')




````

````python
###Start code here
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df2=pd.read_csv('data.csv')
data=df2['Average temperature (°F)']
fig, ax=plt.subplots(figsize=(8, 8))
plot3=sns.distplot(data, kde = True, bins=20,ax=ax)
plot3.set_title('Average temperature distribution')
#ax.set_xlabel('Average temperature')
#plot3=sns.kdeplot(data=data,color='blue')
````

````python
###Start code here
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df3=pd.read_csv('data.csv')
data=df3['Maximum pressure']
plt.figure(figsize=(8,8))
plot4=sns.violinplot(x=data,gridsize=100).set_title('Violin plot')


````
