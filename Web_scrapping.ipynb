```python

    
    three_stars=[]
    base_url='https://books.toscrape.com/catalogue/page-{}.html'
    for i in range(1,51):
        scrape_url=base_url.format(i)
        result=requests.get(scrape_url)
        bs=bs4.BeautifulSoup(result.text,'lxml')
        books=bs.select('.product_pod')
        
        for book in books:
            if len(book.select('.star-rating.Three')) !=0:
                book_title=book.select('a')[1]['title']
                three_stars.append(book_title)
                
```
